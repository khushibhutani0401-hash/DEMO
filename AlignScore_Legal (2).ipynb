```python
# creating the dataset extractors for each dataset
from itertools import islice

def get_billsum_pairs(dataset, n=5):
    pairs = []
    for ex in islice(dataset, n):
        pairs.append({
            "dataset": "BillSum",
            "context": ex["text"],
            "claim": ex["summary"]
        })
    return pairs


def get_casehold_pairs(dataset, n=5):
    pairs = []
    for ex in islice(dataset, n):
        correct_claim = ex["endings"][ex["label"]]
        pairs.append({
            "dataset": "CaseHOLD",
            "context": ex["context"],
            "claim": correct_claim
        })
    return pairs


def get_contractnli_pairs(dataset, n=5):
    pairs = []
    for ex in islice(dataset, n):
        pairs.append({
            "dataset": "ContractNLI",
            "context": ex["premise"],
            "claim": ex["hypothesis"]
        })
    return pairs
```

```python
# combining the dataset in the one dataset
all_pairs = []

all_pairs.extend(get_contractnli_pairs(contract_nli, n=500))
all_pairs.extend(get_billsum_pairs(billsum, n=500))
all_pairs.extend(get_casehold_pairs(casehold, n=500))

print("Total examples:", len(all_pairs))
print("Printing the first example")
print(all_pairs[0])
```

```python
# Running the baseline RoBERTa AlignScore
results = []

for item in all_pairs:
    score = scorer.score(
        contexts=[item["context"]],
        claims=[item["claim"]]
    )[0]

    results.append({
        "dataset": item["dataset"],
        "model": "RoBERTa AlignScore baseline",
        "alignscore": score,
        "claim": item["claim"][:300]
    })

print(results)
```

```python
def convert_contractnli_label(label):
    label = str(label).lower()

    if label in ["entailment", "entailed", "0"]:
        return 1   # aligned

    elif label in ["contradiction", "contradictory", "1"]:
        return 0   # not aligned

    elif label in ["neutral", "2"]:
        return 0   # treat neutral as not fully aligned

    else:
        return None


df_contract_train["align_label"] = df_contract_train["label"].apply(convert_contractnli_label)

df_contract_train.sample(10)
```

```python
# creating dataset + dataloader for LegalBERT fine-tuning
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer

MODEL_NAME = "nlpaueb/legal-bert-base-uncased"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
KeyError: 'args'
```

```python
class LegalAlignDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_length=512):
        self.df = dataframe.reset_index(drop=True)
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        context = str(self.df.loc[idx, "context"])
        claim = str(self.df.loc[idx, "claim"])
        label = int(self.df.loc[idx, "align_label"])

        encoding = self.tokenizer(
            context,
            claim,
            truncation=True,
            padding="max_length",
            max_length=self.max_length,
            return_tensors="pt"
        )

        item = {
            "input_ids": encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0),
            "align_label": torch.tensor(label, dtype=torch.long)
        }

        if "token_type_ids" in encoding:
            item["token_type_ids"] = encoding["token_type_ids"].squeeze(0)

        return item
```

```python
from transformers import AutoTokenizer
import tensorflow as tf
import pandas as pd

MODEL_NAME = "nlpaueb/legal-bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
```

```python
texts_context = df_contract_train["context"].astype(str).tolist()
texts_claim = df_contract_train["claim"].astype(str).tolist()
labels = df_contract_train["align_label"].astype(int).tolist()

encodings = tokenizer(
    texts_context,
    texts_claim,
    truncation=True,
    padding=True,
    max_length=512,
    return_tensors="tf"
)

y = tf.constant(labels)
```

```python
from sklearn.model_selection import train_test_split

input_ids = encodings["input_ids"].numpy()
attention_mask = encodings["attention_mask"].numpy()

X_train_ids, X_val_ids, X_train_mask, X_val_mask, y_train, y_val = train_test_split(
    input_ids,
    attention_mask,
    labels,
    test_size=0.2,
    random_state=42,
    stratify=labels
)
```

```python
from transformers import TFAutoModel
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.models import Model

bert = TFAutoModel.from_pretrained(MODEL_NAME)

input_ids_layer = Input(shape=(512,), dtype=tf.int32, name="input_ids")
attention_mask_layer = Input(shape=(512,), dtype=tf.int32, name="attention_mask")

bert_output = bert(
    input_ids=input_ids_layer,
    attention_mask=attention_mask_layer
)

cls_output = bert_output.last_hidden_state[:, 0, :]

x = Dropout(0.2)(cls_output)
output = Dense(1, activation="sigmoid")(x)

legalbert_keras_model = Model(
    inputs=[input_ids_layer, attention_mask_layer],
    outputs=output
)

legalbert_keras_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=2e-5),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

legalbert_keras_model.summary()
```

```python
history = legalbert_keras_model.fit(
    {
        "input_ids": X_train_ids,
        "attention_mask": X_train_mask
    },
    tf.constant(y_train),
    validation_data=(
        {
            "input_ids": X_val_ids,
            "attention_mask": X_val_mask
        },
        tf.constant(y_val)
    ),
    epochs=3,
    batch_size=4
)
```

```python
def legalbert_alignscore(context, claim):
    encoding = tokenizer(
        [context],
        [claim],
        truncation=True,
        padding="max_length",
        max_length=512,
        return_tensors="tf"
    )

    score = legalbert_keras_model.predict({
        "input_ids": encoding["input_ids"],
        "attention_mask": encoding["attention_mask"]
    })

    return float(score[0][0])
```

```python
context = "The court held that the contract was valid and enforceable."
claim = "The contract was valid."

score = legalbert_alignscore(context, claim)
print(score)
```
